```python
import pandas as pd

url = "https://raw.githubusercontent.com/mhjabreel/CharCnn_Keras/master/data/ag_news_csv/train.csv"

df = pd.read_csv(url, header=None)

print(df.head())
print(df.shape)
```

```python
print(df[0].unique())
print(df[0].value_counts())
```

```python
label_map = {
    1: "World",
    2: "Sports",
    3: "Business",
    4: "Sci/Tech"
}

df["category"] = df[0].map(label_map)

print(df[["category", 1]].head())
```

```python
# Create one text column by combining title and description

df["text"] = df[1] + " " + df[2]

# Keep only useful columns

df = df[["text", "category"]]

print(df.head())

print("\nDataset Shape:", df.shape)

```

```python
sample_df = df.sample(n=2000, random_state=42)

print(sample_df.shape)

sample_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

sample_df["label"] = encoder.fit_transform(sample_df["category"])

print(sample_df[["category", "label"]].head())

print("\nClasses:")
print(encoder.classes_)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    sample_df["text"],
    sample_df["label"],
    test_size=0.2,
    random_state=42
)

print("Training Samples:", len(X_train))
print("Testing Samples :", len(X_test))
```

```python
def planner(user_input):

    user_input = user_input.lower()

    if "entity" in user_input or "person" in user_input or "organization" in user_input:
        return "NER"

    elif "news" in user_input or "category" in user_input or "classify" in user_input:
        return "CLASSIFICATION"

    else:
        return "BOTH"


print(planner("Find entities in this text"))
print(planner("Classify this news article"))
print(planner("Analyze this article"))
```

```python
def classify_news(text):

    text = text.lower()

    if any(word in text for word in ["stock", "market", "business", "oil", "finance"]):
        return "Business"

    elif any(word in text for word in ["football", "cricket", "match", "league", "player"]):
        return "Sports"

    elif any(word in text for word in ["technology", "ai", "computer", "software", "iphone"]):
        return "Sci/Tech"

    else:
        return "World"


print(classify_news("Apple released a new AI software"))
print(classify_news("Pakistan won the cricket match"))
print(classify_news("Oil prices increased today"))
```

```python
def agent(user_input):

    plan = planner(user_input)

    result = {}

    # Memory Save
    memory.append(user_input)

    if plan == "NER":

        result["plan"] = "NER"
        result["entities"] = ner(user_input)

    elif plan == "CLASSIFICATION":

        result["plan"] = "CLASSIFICATION"
        result["category"] = classify_news(user_input)

    else:

        result["plan"] = "BOTH"
        result["category"] = classify_news(user_input)
        result["entities"] = ner(user_input)

    return result
```

```python
response = agent(
    "Apple announced a new AI software in London"
)

print(response)
```

```python
memory = []

def planner(user_input):
    user_input = user_input.lower()

    if "entity" in user_input or "person" in user_input or "organization" in user_input:
        return "NER"
    elif "news" in user_input or "category" in user_input or "classify" in user_input:
        return "CLASSIFICATION"
    else:
        return "BOTH"

def classify_news(text):
    text = text.lower()

    if any(word in text for word in ["stock","market","business","oil","finance"]):
        return "Business"
    elif any(word in text for word in ["football","cricket","match","league","player"]):
        return "Sports"
    elif any(word in text for word in ["technology","ai","computer","software","iphone"]):
        return "Sci/Tech"
    else:
        return "World"

def agent(user_input):
    plan = planner(user_input)

    result = {}

    memory.append(user_input)

    if plan == "CLASSIFICATION":
        result["plan"] = "CLASSIFICATION"
        result["category"] = classify_news(user_input)
    else:
        result["plan"] = plan

    return result

print(agent("Pakistan won the cricket match yesterday"))
```
